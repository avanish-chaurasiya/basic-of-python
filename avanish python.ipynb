```python
#factorial by using recursion

def fact(n):
  if(n==0 or n==1):

    return 1
  else:
    return n* fact(n-1)

fact(8)
```

```python
#problem 1 write a recursive function to calculate the sum of first n number

def sum(n):
  if (n==1):
    return 1
  else:
    return n + sum(n-1)

sum(10)
```

```python
#problem 2 write a recursive function to print all element in a list

def print_list(list,idx=0):
  if(idx == len(list)):
      return
  print(list[idx])
  print_list(list,idx+1)

name = ["avanish","bipin","Soni","Moni"]
print_list(name)
```
